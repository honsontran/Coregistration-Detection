```python
# Currently tested and working with labelme 3.14.1
# install this by typing in cmd pip install opencv-python (which should have 4.1.0 as version #)
import cv2

import json
import numpy as np
from matplotlib import pyplot as plt
```

```python
import glob 

# Make sure masks and and labels are in one folder together. Reference dirs below.

# Mac/Linux
# labels_dir = '/honson/catheterLabels/set1'
# mask_dir = '/honson/catheterLabels/set1/masks/1'

# Windows
labels_dir = 'D:\\Users\\iFai1\\Desktop\\mask-generation\\images'
mask_dir = 'D:\\Users\\iFai1\\Desktop\\mask-generation\\masks\\0'

# labels_dir = labels_dir.replace("\\", "/")
# mask_dir = mask_dir.replace("\\", "/")

fns = glob.glob(labels_dir+'\\*.json')
```

```python
# This can be run to see if your directories are being detected :D
import os 
def printDir(rootDir): 
    for lists in os.listdir(rootDir): 
        path = os.path.join(rootDir, lists) 
        print(path) 
#         if os.path.isdir(path): 
#             Test2(path)
            
printDir(labels_dir)
```

```python
# This label can be set to a "category" of labels. If you are only masking for 1 category, naming is arbitrary.
# Just make sure this label corresponds with mask_dir in the cell above.
my_label='0'

for fn in fns:
    
    label_fn = fn
    
    f = open(label_fn) 
    data = json.load(f)
    
    shapes = []

    for i in data['shapes']:

        label = i['label']
        poly = np.asarray(i['points'], np.int32)
        area = cv2.contourArea(poly) 

        shapes.append((label, area, poly))
                    
    shape_list = sorted(shapes, key=lambda x: x[1], reverse=True)
    
    # Change the resolution to the width and height of the image to correctly create the mask
    w, h = 512, 512
    gray = np.zeros((h,w), np.uint8)
    gray[:,:] = 0

    for i in range(len(shape_list)): 
        if shape_list[i][0] == my_label:
            print('Extracting', shape_list[i][0])
            gray = cv2.fillPoly(gray, [shape_list[i][2]], 255)
    
    # Windows
    filename = fn.split('\\')[-1][:-4]+'jpg'
    
    # Mac/Linux
    # filename = fn.split('/')[-1][:-4]+'jpg'

    print(filename)
    
    # Windows
    print('Writing to: ' + mask_dir+'\\'+filename)
    cv2.imwrite(mask_dir+'\\'+filename, gray)
    
    # Mac/Linux
    # print('Writing to: ' + mask_dir+'/'+filename)
    # cv2.imwrite(mask_dir+'/'+filename, gray)
```
